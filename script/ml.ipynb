```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load features and labels

file_truth = "../data/raw_data/example_ground_truth.csv"
file_border = "./border.csv"
file_ratio = "./ratio.csv"
file_color = "./color.csv"

df_truth = pd.read_csv(file_truth)
df_border = pd.read_csv(file_border)
df_ratio = pd.read_csv(file_ratio)
df_color = pd.read_csv(file_color)
df_color["color_range"] = df_color["color_max"] - df_color["color_min"]

# Combine the features and labels
df = df_truth.drop(["seborrheic_keratosis"],axis=1)

df = pd.merge(df, df_border, on="image_id")
df = pd.merge(df, df_ratio, on="image_id")
df = pd.merge(df, df_color, on="image_id")

print(df.shape)
df
```

```python
from sklearn.model_selection import train_test_split

df2 = df
# noisy fetures
noise = np.random.RandomState(42).uniform(0,0.1, size=(df2.shape[0], 20))

# add the noisy data to the information features
# X = np.hstack((df2[['area', 'perimeter', 'circularity', 'ratio', 'color_variance', "color_min", "color_max", "color_median", "color_mean", "color_range"]], noise))
X = np.hstack((df2[['area','ratio', 'circularity', 'color_variance']], noise))
y = df2["melanoma"]

# split dataset to select feature and evaluate the classifier
X_dev, X_test, y_dev, y_test = train_test_split(X,y, stratify= y, random_state=0)

X_train, X_val, y_train, y_val = train_test_split(X_dev,y_dev, stratify= y_dev)
print(X_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

# Select features 

X_train2 = X_train[:, [0,1,2,3]] 

# Train a classifier
knn1 = KNeighborsClassifier(n_neighbors=1)
knn1trained = knn1.fit(X_train2, y_train)

knn2 = KNeighborsClassifier(n_neighbors=5)
knn2trained = knn2.fit(X_train2, y_train)

tree1 = DecisionTreeClassifier()
tree1trained = tree1.fit(X_train2, y_train)
```

```python
X_val2 = X_val[:, [0,1,2,3]]

y_val_knn1 = knn1trained.predict(X_val2)
y_val_knn2 = knn2trained.predict(X_val2)
y_val_tree1 = tree1trained.predict(X_val2)

# Accuracy - often used in ML but not suitable for medical imaging in general
print(np.sum(y_val_knn1 == y_val)/np.size(y_val)*100)
print(np.sum(y_val_knn2 == y_val)/np.size(y_val)*100)
print(np.sum(y_val_tree1 == y_val)/np.size(y_val)*100)
```

```python
from sklearn.metrics import accuracy_score

acc_knn1 = accuracy_score(y_val, y_val_knn1)
acc_knn2 = accuracy_score(y_val, y_val_knn2)
acc_tree1 = accuracy_score(y_val, y_val_tree1)

print(acc_knn1, acc_knn2, acc_tree1)
```

```python
from sklearn.metrics import roc_auc_score

auc1 = roc_auc_score(y_val, y_val_knn1)
auc2 = roc_auc_score(y_val, y_val_knn2)
auc3 = roc_auc_score(y_val, y_val_tree1)

print(auc1,auc2, auc3)
```

```python
X_test = X_test[:, [0,1,2,3]]

y_test_knn1 = knn2trained.predict(X_test)

acc_test = accuracy_score(y_test, y_test_knn1)
auc_test = roc_auc_score(y_test, y_test_knn1)

print(acc_test)
print(auc_test)
```
