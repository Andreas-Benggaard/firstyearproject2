```python
from skimage.segmentation import slic, mark_boundaries

# Divide the pixels into segments (segment = piece of continuous color in the image)
segments_slic = slic(im, n_segments=250, compactness=10, sigma=1, start_label=1)

fig, ax = plt.subplots(2, 1, figsize=(10, 10), sharex=True, sharey=True)

ax[0].imshow(segments_slic)

ax[1].imshow(mark_boundaries(im, segments_slic))

ax[1].set_title('SLIC')


for a in ax.ravel():
    a.set_axis_off()
  

plt.tight_layout()
plt.show()
```

```python
print((min(np.where(mask2==1.0)[1])))

lesiondimensions=np.where(mask2==1.0)

minx= min(lesiondimensions[0])
maxx= max(lesiondimensions[0])
miny=min(lesiondimensions[1])
maxy=max(lesiondimensions[1])

lesion=im[minx:maxx,miny:maxy]

segmented_lesion=slic(lesion,n_segments=10, compactness=10, sigma=1, start_label=1)
plt.imshow(mark_boundaries(lesion,segmented_lesion))

all_averages=[]
for (i, segVal) in enumerate(np.unique(segmented_lesion)):
    mask=np.zeros(lesion.shape[:2],dtype='uint8')
    mask[segmented_lesion==segVal]=255
    merged=cv2.bitwise_and(lesion, lesion, mask = mask)

    red=0
    green=0
    blue=0
    pixels=0
    for x in merged:
        for pixel in x:
            red+=pixel[0]
            green+=pixel[1]
            blue+=pixel[2]
            if pixel[0]!=0 and pixel[1]!=0 and pixel[2]!=0:
                pixels+=1

    
    superpixel_average=[red/pixels, green/pixels, blue/pixels]
    all_averages.append(superpixel_average)


print(all_averages)
```
